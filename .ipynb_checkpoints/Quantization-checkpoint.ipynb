# Introduction to Quantization


### 📦 Model Quantization with TensorFlow & Keras

Quantization is a powerful model optimization technique that reduces the **precision** of the weights and activations used in a neural network.  
This allows deep learning models to be deployed efficiently on **edge devices**, **mobile phones**, and **microcontrollers** where memory and computation are limited.

In this notebook, we demonstrate the process and effects of:
- Post-Training Quantization (PTQ)
- Quantization-Aware Training (QAT)

We will compare models in terms of **size**, **accuracy**, and **inference performance**.


#  Quantization Types


## 🔍 Types of Quantization

| Type | Description | Retraining Needed | Pros | Cons |
|------|-------------|-------------------|------|------|
| **PTQ** (Post-Training Quantization) | Quantizes weights/activations *after* training | ❌ | Simple, fast | Possible accuracy drop |
| **QAT** (Quantization-Aware Training) | Simulates quantization *during* training | ✅ | High accuracy retention | Slower training |
| **Dynamic Range Quantization** | Only weights are quantized; activations done at runtime | ❌ | Fastest & easiest | Lower compression |
| **Float16 Quantization** | Converts model to float16 format | ❌ | Fast GPU inference | Less compression than int8 |


## 🎯 Goal

To analyze the **effect of quantization and compression** on:
- Model accuracy (train/test)
- Inference latency and throughput
- Model loading time
- File size and deployment readiness

```python
import tensorflow
#Quantization
import tensorflow_model_optimization.python.core.keras.compat import keras
import tensorflow_model_optimization as tfmot
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from keras.metrics import Precision,Recall,F1Score
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
np.unique(y_train)
```

```python
x_train=x_train.reshape(60000,28*28)
x_test=x_test.reshape(10000,28*28)
```

```python
model=keras.Sequential()
model.add(keras.layers.Input(shape=(784,)))
model.add(keras.layers.Dense(units=100,activation='relu'))
model.add(keras.layers.Dense(units=100,activation='relu'))
model.add(keras.layers.Dense(units=100,activation='relu'))
model.add(keras.layers.Dense(units=10,activation='softmax'))
model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
model.fit(x_train,y_train,epochs=5,batch_size=256,validation_data=[x_test,y_test])
```

# PTQ - Compression

### Full Form: Post-Training Quantization
### Use: Compresses and speeds up a pre-trained model after training, without retraining.

```python
model1=keras.models.load_model(r"normal.h5")
#covert model
converter=tensorflow.lite.TFLiteConverter.from_keras_model(model1) #loadmodel
converter.optimizations=[tensorflow.lite.Optimize.DEFAULT]
tflite_quant_model=converter.convert()

#save the model
with open((r"normal_compressed.tflite"),mode="wb") as file:
    file.write(tflite_quant_model)
```

# PTQ - Compression + Quantization

```python
# Load the model --> Netron app
```

```python
model2=keras.models.load_model(r"normal.h5")

#quantize the loded model
quantize_model=tfmot.quantization.keras.quantize_model
quantize_model=quantize_model(model2)

#convert the model
converter=tensorflow.lite.TFLiteConverter.from_keras_model(model2)
converter.optimizations=[tensorflow.lite.Optimize.DEFAULT]
tflite_quant_model=converter.convert()

#save the model

with open((r"normal_quantized_compressed.tflite"),mode="wb") as file:
    file.write(tflite_quant_model)
```

> `Compression` only reduces the file size **without changing the data type**, while

> `Compression + Quantization` reduces both the **size and converts the data type (e.g., from float32 to int8)** for more efficient computation.

# Quantization Aware Training

* Normal Approach : create a model --> compile the model --> fit the model -- graph (TA and VA) | (TL and VL)
* Quantization Aware Approach :create a model --> quantize the model --> compile the model --> fit the model

### Normal Approach : create a model --> compile the model --> fit the model -- graph (TA and VA) | (TL and VL)

```python
import time
```

```python
start=time.time()
model=keras.Sequential()
model.add(keras.layers.Input(shape=(784,)))
model.add(keras.layers.Dense(units=100,activation='relu'))
model.add(keras.layers.Dense(units=100,activation='relu'))
model.add(keras.layers.Dense(units=100,activation='relu'))
model.add(keras.layers.Dense(units=10,activation='softmax'))
#model.summary()

model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
history = model.fit(x_train,y_train,epochs=5,batch_size=256,validation_data=[x_test,y_test])
end=time.time()

total_time = end-start
print(f"total time taken is {total_time:.3f} seconds")
```

```python
# Create a figure with 1 row and 2 columns
plt.figure(figsize=(12, 5))

# ------------------ Plot 1: Accuracy ------------------
plt.subplot(1, 2, 1)
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("Train vs Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend(["Train", "Validation"])

# ------------------ Plot 2: Loss ------------------
plt.subplot(1, 2, 2)
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Train vs Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend(["Train", "Validation"])

# ------------------ Show Plot ------------------
plt.tight_layout()
plt.show()
```
